```python
import pickle
from pymongo import MongoClient
from werkzeug.security import generate_password_hash
```

```python
# Load MongoDB
client = MongoClient('mongodb+srv://triskaTA:Triskalulus2025@cluster0.egskx8p.mongodb.net/')
db = client['DATA_SKRIPSI_API']
user_collection = db['data_users']
```

```python
# Load user_encoder.pkl
with open('user_encoder.pkl', 'rb') as f:
    user_encoder = pickle.load(f)
```

```python
db_scrape = client['DATA_SKRIPSI']
comment_collection = db_scrape['part1_data_comment']
```

```python
# Ambil daftar user
usernames_lama = comment_collection.distinct("username")
usernames_lama
```

```python
len(usernames_lama)
```

```python
# Filter username yang sudah ada di user_encoder
usernames_lama_filtered = [u for u in usernames_lama if u in user_encoder.classes_]
```

```python
usernames_lama_filtered
```

```python
len(usernames_lama_filtered)
```

```python
# Transform hanya username yang sudah ada
user_ids = user_encoder.transform(usernames_lama_filtered)
```

```python
print("User IDs setelah filtering:", user_ids)
```

```python
# Simpan user lama ke MongoDB dengan user_id dan password default "12345"
for username, user_id in zip(usernames_lama, user_ids):
    user_collection.insert_one({
        "username": username,
        "user_id": int(user_id),
        "password": generate_password_hash("12345")
    })

print("User lama berhasil dimasukkan ke MongoDB dengan user_id!")
```

```python
# Ambil semua user tanpa user_id
users = user_collection.find({"user_id": {"$exists": False}})
```

```python
print(users)
```

```python
# Update user lama dengan user_id
for user in users:
    username = user["username"]
    if username in user_encoder.classes_:
        user_id = user_encoder.transform([username])[0]
        user_collection.update_one({"_id": user["_id"]}, {"$set": {"user_id": user_id}})
```
