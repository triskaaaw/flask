```python
from pymongo import MongoClient
import pandas as pd
import numpy as np
```

```python
client = MongoClient('mongodb+srv://triskaTA:Triskalulus2025@cluster0.egskx8p.mongodb.net/')
```

```python
db = client['DATA_SKRIPSI_API']
db_raw = client['DATA_SKRIPSI']
```

```python
resep_collection = db['data_resep']
data_resep = list(resep_collection.find())
df_resep = pd.DataFrame(data_resep)
df_resep.head()
```

```python
data_komen = db_raw['part1_data_comment']
data_komen = list(data_komen.find())
df_komen = pd.DataFrame(data_komen)
df_komen.head()
```

```python
data_rating = df_resep[df_resep['ratingValue']== 5]
data_rating.head()
```

```python
data_rating.info
```

```python
# 1. Hitung jumlah komentar per title
jumlah_komen = df_komen.groupby('title').size().reset_index(name='jumlah_komentar')

# 2. Gabungkan dengan data_rating berdasarkan title
data_rating_komentar = pd.merge(data_rating, jumlah_komen, on='title', how='inner')

# 3. Ambil 5 teratas berdasarkan jumlah komentar
top_5_resep = data_rating_komentar.sort_values(by='jumlah_komentar', ascending=False).head(5)

# 4. Tampilkan
top_5_resep[['idRecipe','title', 'image_url', 'ratingValue', 'jumlah_komentar', 'time']]
```
